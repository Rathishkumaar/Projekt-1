```python
import wbgapi as wb
import plotly.express as px

# Retrieve GDP series data
data = wb.series.list(q="GDP")

# Store the ids and titles in lists
data_list = []
choice = 0
for id in data:
    id_title = id["id"]
    data_list.append(id_title)
    print(str(choice) + ": " + str(id["value"]))
    choice = choice + 1

# Prompt user to select a database
selection = int(input("Please enter the number of the database you would like to chart: "))

# Retrieve data for selected series
df = wb.data.DataFrame(data_list[selection], country_list, range(2024), index="time", numericTimeKeys=True, labels=True)

# Create and display the line chart
fig = px.line(df, x="Time", y="USA")
fig.update_layout(xaxis=dict(autorange="reversed"))
fig.show()
```
