```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
sns.barplot(x="Ladder score", y="log_GDP", data=merged_df)
```

```python

# Filter the DataFrame to include only the desired range of indexes
filtered_df = sorted_df[(sorted_df.index >= 0) & (sorted_df.index <= 5)]

# Plot the filtered data
sns.barplot(x="Country name", y="log_GDP", data=filtered_df)
```

```python
# Assuming merged_df is your DataFrame
subset_df = pd.concat([sorted_df.iloc[:5], sorted_df.iloc[-5:]])
sns.barplot(x="Country name", y="log_GDP", data=subset_df)
```
